```python
# Rank sütununda Null, NaN ve 'Bilinmiyor' yazanları silme
df_cleaned = df[~df['Rank'].isin([None, 'NaN', 'Bilinmiyor'])].dropna(subset=['Rank'])
```

```python
df_cleaned.info()
```

```python
df_cleaned.tail(5)
```

```python
df_transformed = pd.DataFrame({
    "Question": df_cleaned["Name"].apply(lambda x: f"{x} kimdir?"),
    "Answer": df_cleaned.apply(
        lambda row: f"{row['Name']} Fırat Üniversite kurumunda {row['Rank']} derecesinde çalışmaktadır. Çalışma alanı ise şunlardır: {row['Work Field']}", axis=1
    )
})
```
